```python
%matplotlib inline 

import sys
sys.path.append('../')
sys.path.append('../../stackedgp_src')
import numpy as np
from __future__ import division
from mpl_toolkits.mplot3d import axes3d
import GPy, sys, matplotlib
import matplotlib.pyplot as plt
from stackedGPNetwork import StackedGPNetwork
from toyUtil import ToyUtil
```

```python
def generateX(nsamples, start=-5, end=5):
    x = np.random.uniform(start,end,(nsamples,1))
    return x

def getT(x, xid):
    if xid == 1:
        return np.sin(x)
    else:
        return np.sin(x)

def getA(t1, t2):
    return t1*t2
```

```python
#generate X,T,A for training
ntraining = 20
x1_train = generateX(ntraining)
x2_train = generateX(ntraining)

t1_train = getT(x1_train, 1)
t2_train = getT(x2_train, 2)

# X, Y, Z = axes3d.get_test_data(0.05)
t1_r_train = np.repeat(t1_train, ntraining, axis=1)
t2_r_train = np.repeat(t2_train, ntraining, axis=1).T

x1_r_train = np.repeat(x1_train, ntraining, axis=1)
x2_r_train = np.repeat(x2_train, ntraining, axis=1).T

a_train = getA(t1_r_train, t2_r_train)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# ax.scatter(x[:,0:1],x[:,1:2], a_train)
ax.scatter(x1_r_train,x2_r_train, a_train)
plt.show()
```

```python
#generate X,T,A for training
ntesting = 20
x1_test = generateX(ntesting, start=-5, end=5)
x2_test = generateX(ntesting, start=-5, end=5)

t1_test = getT(x1_test, 1)
t2_test = getT(x2_test, 2)

t1_r_test = np.repeat(t1_test, ntesting, axis=1)
t2_r_test = np.repeat(t2_test, ntesting, axis=1).T

x1_r_test = np.repeat(x1_test, ntesting, axis=1)
x2_r_test = np.repeat(x2_test, ntesting, axis=1).T

a_test = getA(t1_r_test, t2_r_test)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1_r_test,x2_r_test, a_test)
plt.show()
```

```python
#Train StackedGP layer
layer2GPinput = np.hstack((t1_r_train.reshape(-1,1),t2_r_train.reshape(-1,1)))

stackedNetwork = StackedGPNetwork(2)
stackedNetwork.createNewNode(1,layer2GPinput,a_train.reshape(-1,1), normalize=True, useGPU=False)
stackedNetwork.createNewNode(0,x1_train,t1_train, normalize=True, useGPU=False)
stackedNetwork.createNewNode(0,x2_train,t2_train, normalize=True, useGPU=False)
```

```python
#optimize the stacked GP
stackedNetwork.optimize(numoptimizationtrails=5)
```

```python
stackedNetwork.getNode(1,0)
```

```python
print 'Testing Analytical approach on training samples....'
testinginput = np.concatenate((x1_r_train.reshape(-1,1), x2_r_train.reshape(-1,1)), axis=1)
anlyt_mean_train, var = stackedNetwork.predict(testinginput, covoption=2, jitter=1e-1)
anlyt_mean_train = anlyt_mean_train[-1]
var = var[-1]
anlyt_mean_train = anlyt_mean_train.reshape(-1,ntraining)

ToyUtil.plottingScatter3D(x1_r_train,'x1_train',x2_r_train,'x2_train', a_train,'Actual_Data_train',anlyt_mean_train,
                       'Predicting_Analytical_mean', savefname='toy_12_3d_Analytical_training.png')

anlyt_train_err, anlyt_train_rms = ToyUtil.getRMS(a_train, anlyt_mean_train)

print 'Min Error = ', anlyt_train_err.min(), ' Max Error = ', anlyt_train_err.max()
print 'anlyt_train_RMS = ', anlyt_train_rms
```

```python
print 'Testing Analytical approach on testing samples....'
testinginput = np.concatenate((x1_r_test.reshape(-1,1), x2_r_test.reshape(-1,1)), axis=1)
anlyt_mean_test, var_test = stackedNetwork.predict(testinginput, covoption=1, jitter=0)
anlyt_mean_test = anlyt_mean_test[-1]
var_test = var_test[-1]
anlyt_mean_test = anlyt_mean_test.reshape(-1,ntesting)

ToyUtil.plottingScatter3D(x1_r_test,'x1_test',x2_r_test,'x2_test', a_test,'Actual_Data_test', anlyt_mean_test,
                       'Predicting_Analytical_mean', savefname='toy_12_3d_Analytical_testing_v01.png')

anlyt_test_err, anlyt_test_rms = ToyUtil.getRMS(a_test, anlyt_mean_test)

print 'Min Error = ', anlyt_test_err.min(), ' Max Error = ', anlyt_test_err.max()
print 'anlyt_test_RMS = ', anlyt_test_rms
```
